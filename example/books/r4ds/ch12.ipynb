```python
from hea.tidy import col
from hea.data import data
```

```python
flights = data("flights")
```

### 12.2

```python
(
    flights.
        filter(
            (col("dep_time") > 600) & \
            (col("dep_time") < 2000) & \
            (abs(col("arr_delay")) < 20))
)
```

```python
(
    flights
        .mutate(
            daytime = (col("dep_time") > 600) & (col("dep_time") < 2000),
            approx_ontime = abs(col("arr_delay")) < 20,
            _keep = "used"
        )
)
```

```python
(
    flights
        .mutate(
            daytime = (col("dep_time") > 600) & (col("dep_time") < 2000),
            approx_ontime = abs(col("arr_delay")) < 20,
        ).filter(col("daytime") & col("approx_ontime"))
)
```

```python
flights.filter(col("dep_time").is_nan())
```

```python
flights.filter(col("dep_time").is_null())
```
