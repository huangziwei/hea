### 3.6

```python
batters = (
    data("batting", package="Lahman")
        .group_by("playerID")
        .summarize(
            performance=col("H").sum() / col("AB").sum(),
            n = col("AB").sum()
        )
)
batters
```

```python
(
    batters
        .filter(col("n")>100)
        .ggplot(x="n", y="performance")
        .geom_point(alpha=0.1)
        .geom_smooth(se=False)
)
```

```python
batters.arrange(desc("performance"))
```
