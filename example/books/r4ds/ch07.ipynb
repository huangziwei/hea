### 7.2.1

```python
students = read_csv("https://pos.it/r4ds-students-csv", null_values=["N/A", ""])
students
```

```python
from hea import factor, parse_number, if_else
students.rename(
        student_id = "Student ID",
        full_name = "Full Name",
        meal_plan= "mealPlan",
        age = "AGE",
    ).mutate(
        meal_plan=factor("meal_plan"),
        age = parse_number(if_else(col("age") == "five", "5", col("age")))
    )
```

```python
students.clean_names()
```

```python
(
    students
        .clean_names()
        .mutate(meal_plan = factor("meal_plan"))
)
```

```python
students = (
    students
        .clean_names()
        .mutate(
            meal_plan = factor("meal_plan"),
            age = parse_number(if_else(col("age") == "five", "5", col("age")))
        )
        
)
students
```
