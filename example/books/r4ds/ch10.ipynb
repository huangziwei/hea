```python
from hea import data, col
```

### 10.3

```python
diamonds = data("diamonds")
smaller = diamonds.filter(col("carat") < 3)
```

```python
p1 = diamonds.ggplot(x="carat").geom_histogram(binwidth=0.5)
p2 = smaller.ggplot(x="carat").geom_histogram(binwidth=0.01)
p3 = diamonds.ggplot(x="y").geom_histogram(binwidth=0.5)
p4 = diamonds.ggplot(x="y").geom_histogram(binwidth=0.5).coord_cartesian(ylim=[0, 50])

(p1 | p2) / (p3 | p4)
```

### 10.5

```python
from hea.ggplot import after_stat


p1 = diamonds.ggplot(x="price").geom_freqpoly(color="cut", binwidth=500, linewidth=0.75)
p2 = diamonds.ggplot(x="price", y=after_stat("density")).geom_freqpoly(color="cut", binwidth=500, linewidth=0.75)
p1 | p2
```

```python
diamonds.ggplot(x="cut", y="price").geom_boxplot()
```

```python
from hea import fct_reorder

mpg = data("mpg", "ggplot2")
mpg.ggplot(x=fct_reorder("class", "hwy", "median"), y="hwy").geom_boxplot()
```

```python
mpg.ggplot(x="hwy", y=fct_reorder("class", "hwy", "median")).geom_boxplot()
```

```python
diamonds.ggplot(x="cut", y="color").geom_count()
```

```python
diamonds.count("color", "cut").ggplot(x="color", y="cut").geom_tile(fill="n")
```

```python
p1 = smaller.ggplot(x="carat", y="price").geom_point()
p2 = smaller.ggplot(x="carat", y="price").geom_point(alpha=0.01)
p1 | p2 
```

```python
p1 = smaller.ggplot(x="carat", y="price").geom_bin2d()
p2 = smaller.ggplot(x="carat", y="price").geom_hex()
p1 | p2
```

```python
from hea import cut_width
p1 = smaller.ggplot(x="carat", y="price").geom_boxplot(group=cut_width("carat", 1))
p2 = smaller.ggplot(x="carat", y="price").geom_boxplot(group=cut_width("carat", 0.1))

p1 + p2 
```
