```python
from hea import data
```

### 19.2

```python
flights = data("flights")
flights
```

```python
airlines = data("airlines")
airlines
```

```python
airports = data("airports")
airports
```

```python
planes = data('planes')
planes
```

```python
weather = data("weather")
weather
```

```python
from hea.tidy import col
planes.count("tailnum").filter(col("n") > 1)
```

```python
weather.count("time_hour", "origin").filter(col("n") > 1)
```

```python
planes.filter(col("tailnum").is_null())
```

```python
weather.filter(col("time_hour").is_null() | col("origin").is_null())
```

```python
flights.count("time_hour", "carrier", "flight").filter(col("n") > 1)
```

```python
airports.count("alt", "lat").filter(col("n") > 1)
```

```python
from hea.tidy import row_number
flights2 = flights.mutate(id=row_number(), _before=1)
flights2
```

### 19.3

```python
flights2 = flights.select("year", "time_hour", "origin", "dest", "tailnum", "carrier")
flights2
```

```python
flights2.left_join(airlines)
```

```python
flights2.left_join(
    weather.select("origin", "time_hour", "temp", "wind_speed")
)
```

```python
flights2.left_join(
    planes.select("tailnum", "type", "engines", "seats")
)
```

```python
flights2.filter(col("tailnum") == "N3ALAA").left_join(
    planes.select("tailnum", "type", "engines", "seats")
)
```

```python
flights2.left_join(planes)
```

```python
from hea.tidy import join_by
flights2.left_join(planes, join_by("tailnum"))
```

```python
flights2.left_join(airports, join_by(col("dest") == col("faa")))
```

```python
flights2.left_join(airports, join_by(col("origin") == col("faa")))
```

```python
airports.semi_join(flights2, join_by(col("faa") == col("origin")))
```

```python
airports.semi_join(flights2, join_by(col("faa") == col("dest")))
```

```python
flights2.anti_join(airports, join_by(col("dest") == col("faa"))).distinct("dest")
```

```python
flights2.anti_join(planes, join_by("tailnum")).distinct("tailnum")
```
