### 1.2.4

```python
(
    penguins
        .ggplot(
            x="flipper_length_mm", 
            y="body_mass_g",
            color="species"
        )
        .geom_point()
        .geom_smooth(method="lm")
)
```

```python
penguins.ggplot(x="species").geom_bar()
```

```python
from hea.ggplot import fct_infreq
penguins.ggplot(x=fct_infreq("species")).geom_bar()
```

```python
p1 = penguins.ggplot(x ="body_mass_g").geom_histogram(binwidth = 20)
p2 = penguins.ggplot(x ="body_mass_g").geom_histogram(binwidth = 200)
p3 = penguins.ggplot(x ="body_mass_g").geom_histogram(binwidth = 2000)
p1 + p2 + p3
```

```python
penguins.ggplot(x="body_mass_g").geom_density()
```

```python
penguins.ggplot(x="species", y="body_mass_g").geom_boxplot()
```

```python
p1 = penguins.ggplot(x="body_mass_g", color="species").geom_density()
p2 = penguins.ggplot(x="body_mass_g", color="species", fill="species").geom_density(alpha=0.5)
p1 + p2
```

```python
p1 = penguins.ggplot(x="island", fill="species").geom_bar()
p2 = penguins.ggplot(x="island", fill="species").geom_bar(position="fill").labs(y="proportion")
p1 | p2
```

```python
penguins.ggplot(x="flipper_length_mm", y="body_mass_g").geom_point(color="species", shape="species").facet_wrap("~island")
```
