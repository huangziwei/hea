### 5.2

```python
table1 = data("table1", package="tidyr")
table2 = data("table2", package="tidyr")
table3 = data("table3", package="tidyr")
```

```python
table1.mutate(rate=col("cases") / col("population") * 10_000)
```

```python
(
    table1
        .group_by("year")
        .summarize(total_cases=col("cases").sum())
)
```

```python
(
    table1
        .ggplot(x="year", y="cases")
        .geom_line(group="country", color="grey50")
        .geom_point(color="country", shape="country")
        .scale_x_continuous(breaks=[1999, 2000])
)
```

### 5.3.1

```python
billboard = data("billboard")
billboard
```

```python
from hea import selectors
billboard.pivot_longer(
    cols=selectors.starts_with("wk"),
    names_to="week",
    values_to="rank",
)
```

```python
billboard.pivot_longer(
    cols=selectors.starts_with("wk"),
    names_to="week",
    values_to="rank",
    values_drop_na=True,
)
```

```python
from hea.dtypes import Int64
billboard_longer = (
    billboard.pivot_longer(
        cols=selectors.starts_with("wk"),
        names_to="week",
        values_to="rank",
        values_drop_na=True,
        names_prefix="wk",
    ).mutate(
        week=col("week").cast(Int64)
    )
)
```

```python
(
    billboard_longer
        .ggplot(x="week", y="rank", group="track")
        .geom_line(alpha=0.25)
        .scale_y_reverse()
)
```

### 5.3.3

```python
who2 = data("who2", "tidyr")
who2
```

```python
(
    who2
        .pivot_longer(
            cols=~who2["country":"year"],
            names_to=["diagnosis", "gender", "age"],
            names_sep="_",
            values_to="count",
        )
)
```

### 5.3.4

```python
household = data("household", "tidyr")
household
```

```python
(
    household
        .pivot_longer(
            cols=~household["family"],
            names_to=[".value", "child"],
            names_sep="_",
            values_drop_na=True,
        )
)
```
