## Chapter 3 of *Linear Models with R* (3e, 2025)

```python
from hea.R import *
from hea.plot import *
from hea import data, lm
from hea.compare import anova
```

### 3.2

```python
gala = data("gala")

m0 = lm('Species ~ Area + Elevation + Nearest + Scruz + Adjacent', gala)
m1 = lm('Species ~ 1', gala)
anova(m1, m0)
```

```python
m2 = lm('Species ~ Elevation + Nearest + Scruz + Adjacent', gala)
anova(m2, m0)
```

```python
summary(m0)
```

```python
summary(lm("Species ~ Area", gala))
```

```python
m3 = lm("Species ~ Elevation + Nearest + Scruz", gala)
anova(m3, m0)
```

```python
m4 = lm('Species ~ I(Area+Adjacent)+ Elevation + Nearest + Scruz', gala)
anova(m4, m0)
```

```python
m5 = lm("Species ~ Area + offset(0.5 * Elevation) + Nearest + Scruz + Adjacent", gala)
anova(m5, m0)
```
